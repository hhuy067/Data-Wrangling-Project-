```python
import pandas as pd
import numpy as np

#load file
sales = pd.read_csv('sales.csv')
prices = pd.read_csv('prices.csv')
```

```python
#convert ordered_at, updated_at to datetime object (using pd.to_datetime)
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
prices['updated_at'] = pd.to_datetime(prices['updated_at'])
```

```python
sales = sales.sort_values(by='ordered_at')
sales
```

```python
prices = prices.sort_values(by='updated_at')
prices
```

```python
tong_hop = pd.merge_asof(sales, prices, left_on='ordered_at', right_on='updated_at', by='product_id', direction='nearest')
```

```python
tong_hop['gia_ban'] = np.where(tong_hop['ordered_at'] >= tong_hop['updated_at'],
                                         tong_hop['new_price'], tong_hop['old_price'])
tong_hop
```

```python
tong_hop['doanh_thu'] = tong_hop['quantity_ordered'] * tong_hop['gia_ban']
tong_doanh_thu = tong_hop.groupby(['product_id'], as_index=False)['doanh_thu'].sum()
tong_doanh_thu
```
